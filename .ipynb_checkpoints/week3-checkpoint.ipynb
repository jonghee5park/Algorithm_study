```python
#시간초과

n = int(input())
cards = list(map(int, input().split()))
m = int(input())
finds = list(map(int, input().split()))
results = []

cards.sort()

for i, find in enumerate(finds):
    results.append(0)
    for card in cards:
        if card == find:
            results[i]+=1
for result in results:
    print(result, end=" ")
```

```python
#성공
from bisect import bisect_left, bisect_right

def CountByRange(nums, value):
    left_idx = bisect_left(nums, value)
    right_idx = bisect_right(nums, value)
    return right_idx - left_idx

n = int(input())
cards = list(map(int, input().split()))
m = int(input())
finds = list(map(int, input().split()))
results = []

cards.sort()


for find in finds:
    results.append(CountByRange(cards, find))
    
for result in results:
    print(result, end=" ")
    
```

```python
import sys
from collections import deque

#%% 입력 변수

n, m = map(int, input().split())
maze = [input() for i in range(n)]

maze = list(map(lambda x: list(x), maze))
dist = [[-1]*m for i in range(n)]

#%% 연산 변수

dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

#%% 알고리즘

q = deque()
q.append((0,0))
dist[0][0] = 0
cnt = 0

while q:
    cur = q.popleft()
    for i in range(4):
        nx = cur[0] + dx [i]
        ny = cur[1] + dy [i]
        if nx >= 0 and nx < n and ny >= 0 and ny < m:
            if dist[nx][ny] < 0 and maze[nx][ny] == '1':
                dist[nx][ny] = dist[cur[0]][cur[1]] + 1
                q.append((nx, ny))
                
cnt = dist[n-1][m-1] + 1
print(cnt)
```

```python
n = int(input())
graph = [[] for _ in range(n+1)]

#%%

for _ in range(n-1):
    n1, n2 = map(int, input().split())
    graph[n1].append(n2)
    graph[n2].append(n1)
    
#%%

q = [1]
result = [i for i in range(n+1)]

#%%
while q:
    x = q.pop(0)
    for nx in graph[x]:
        if result[nx] == nx:
            result[nx] = x
            q.append(nx)
            
for i in range(2, n+1):
    print(result[i])
```
